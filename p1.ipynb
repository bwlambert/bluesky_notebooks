### [Modeling Behavior and Vaccine Hesitancy Using Twitter‑Derived US Population Sentiment during the COVID‑19 Pandemic to Predict Daily Vaccination Inoculations](https://www.mdpi.com/2076-393X/11/3/709)  
**Ref:** 2023, *Vaccines* (MDPI) 

**Data:** ≈18 million US tweets (Jun 2021 – Mar 2022) merged with CDC daily dose counts.  

**Methods:** Fear–hesitancy sentiment indices + multivariate regressions.  

**Key findings:** Twitter‑derived sentiment consistently led first‑dose and booster uptake by 3–5 days (R ≈ 0.77) suggesting social‑media mood is a *leading* rather than lagging indicator of real‑world vaccination behaviour.

## [Public Health Messaging on Twitter During the COVID‑19 Pandemic: Observational Study](https://www.jmir.org/2025/1/e63910)  
**Ref:** 2025, *Journal of Medical Internet Research*  

**Data:** 539 k tweets by 489 verified public‑health experts versus 356 self‑identified “pseudo‑experts,” Jan 2020 – Jan 2021.  

**Methods:** Moral‑foundation & emotion dictionaries; negative‑binomial models of engagement.  

**Key findings (flow → argument → behaviour):** Anger‑ and disgust‑laden posts by pseudo‑experts travelled farther than neutral expert posts, whereas positively framed expert tweets attracted more civil reply threads

```python
hours = pd.read_csv('bluesky_hourly_archives.csv')
hours['hour'] = pd.to_datetime(hours['hour'], format='%Y%m%d%H%M')
hours.set_index('hour', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(hours.index, hours['size'], marker='o', linestyle='-', color='b')
plt.title('Bluesky Archives Size by Hour')
plt.xlabel('Hour')
plt.ylabel('Size (Megabytes)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
sizeinMB = hours['size'].sum()
sizeinGB = sizeinMB / 1024
print(f'Total size of Bluesky archives: {sizeinMB} MB ({sizeinGB:.2f} GB)')
sizeinGB_xz = sizeinGB / 0.1
print(f'Estimated size before xz compression: {sizeinGB_xz:.2f} GB')
```
